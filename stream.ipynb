```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("/home/jerome/Documents/CSV FILE/resume_screening.csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
df.drop(['education_level','field_of_study','certifications_count', 'tech_skills_count','last_job_title', 'has_linkedin_profile'], axis = 1, inplace= True)
```

```python
df
```

```python
num = ['has_open_source_contributions','has_portfolio_or_github','suitable']
for i in df.columns:
    if i not in num:
        df[i] = df[i].fillna(df[i].mean())
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['has_open_source_contributions','has_portfolio_or_github']
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import classification_report
```

```python
X = df.drop('suitable',axis=1)
y = df['suitable']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2, random_state= 25, stratify= y)
```

```python
model = XGBClassifier(random_state= 25)
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
print(classification_report(y_test,pred))
```

```python
xg = XGBClassifier()
param_grid = {
    'n_estimators': [5,10,15],
    'max_depth': [2,3,4],
    'learning_rate': [0.001,0.002,0.003],
    'subsample': [0.8]
    
}
grid_search = GridSearchCV(xg, param_grid= param_grid,cv= 10, scoring= 'f1', verbose= True)
grid_search.fit(X_train,y_train)
print("f1", grid_search.best_score_)
print(grid_search.best_estimator_)
```

```python
xg = grid_search.best_estimator_
pred = xg.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```
