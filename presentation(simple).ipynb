```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv("/home/jerome/Documents/CSV FILE/assign3.csv")
```

```python
df.info()
```

```python
num = df.select_dtypes("number")
for   x in num.columns:
    sns.boxplot(x = num[x])
    plt.show()
```

```python
corr = df.corr()
corr
```

```python
X = df["Property Size (sq. ft.)"]
y = df["Price (USD)"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state= 40)
```

```python
X_train = X_train.to_frame()
X_test = X_test.to_frame()
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)
```

```python
print(f'R2 score is :{r2:.2f}')
print(f'Mean Squared Error is: {mse:.2f}')
```

```python
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error is: {rmse:.2f}")
```
